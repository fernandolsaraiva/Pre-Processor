**Step 8: Fourier transform** (from https://towardsdatascience.com/fourier-transform-for-time-series-292eb887b101)

```python
#Apply fast fourier transform and take absolute values
f=abs(np.fft.fft(df1['Coelho_level']))
#Get the list of frequencies
num=np.size(df1['dt'])
freq = [i / num for i in list(range(num))]
# get the list of spectrums
spectrum=f.real*f.real+f.imag*f.imag
nspectrum=spectrum/spectrum[0]
# plot nspectrum per frequency, with a semilog scale on nspectrum
plt.semilogy(freq,nspectrum)
#Fourier Transform for Time Series. The result is a plot of strength of each frequency.
```

```python
# improve the plot by adding periods in number of weeks rather than  frequency
import pandas as pd
results = pd.DataFrame({'freq': freq, 'nspectrum': nspectrum})
results['period'] = results['freq'] / (1/52)
plt.semilogy(results['period'], results['nspectrum'])
#You will now obtain an x-axis that has the period in weeks:
```

```python
# improve the plot by convertint the data into grouped per week to avoid peaks
results['period_round'] = results['period'].round()
grouped_week = results.groupby('period_round')['nspectrum'].sum()
plt.semilogy(grouped_week.index, grouped_week)
plt.xticks([1, 13, 26, 39, 52])
```

The peak at the end however is the peak at 52 weeks: this is the peak that shows us that a seasonality of period 52 weeks (yearly) is strongly present.

**Step 9: Lyapunov expoents** (adapted from https://blog.abhranil.net/2014/07/22/calculating-the-lyapunov-exponent-of-a-time-series-with-python-code/)

```python
from math import log
def d(series,i,j):
    return abs(series[i]-series[j])
series = []
for i in df1['Coelho_level']:
  series.append(i)
N=len(series)
eps=float(input('Initial diameter bound: '))
dlist=[[] for i in range(N)]
n=0 #number of nearby pairs found
for i in range(N):
    for j in range(i+1,N):
        if d(series,i,j) < eps:
            n+=1
            #print(n)
            for k in range(min(N-i,N-j)):
                dlist[k].append(log(d(series,i+k,j+k)+0.00001))
#a última linha acima pode dar um erro matemático se a diferença for zero
#não será possível calcular o log (0). Por isso acrescentei 0.00001
```

```python
lyapunov = []
for i in range(len(dlist)):
  if len(dlist[i]):
    lyapunov.append(sum(dlist[i])/len(dlist[i]))
plt.figure(figsize=(10, 6))
plt.plot(lyapunov[:15])
#plt.plot(lyapunov)
```
